```python
def salary_statistics(salaries_list):
  minimum = min(salaries_list)
  maximum = max(salaries_list)
  average = sum(salaries_list)/len(salaries_list)
  info = {"minimum salary": minimum, "maximum salary": maximum, "average salary": average}
  return info
```

```python
def job_posting_summary(job_postings):
  number_postings = len(job_postings)
  total = []
  locations = []
  for item in job_postings:
    total.append(item['salary'])
    if item['location'] not in location:
      location.append(item['location'])
  average = sum(total)/len(total)
  new_dict = {'number of postings': number_postings, 'average salary': average, 'locations': location}
  return new_dict
```

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate = 0.1, symbol = '$'):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salalry(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'

```

```python
salary = BaseSalary (100000)
```

```python
salary.show_bonus()
```

```python
class JobPosting1:
  def __init__(self, title, company, location, salary):
    self.title = title
    self.location = location
    self.company = company
    self.salary = salary

  def to_dict(self):
    return {
        'title': self.title,
        'location': self.location,
        'company': self.company,
        'salary': self.salary
    }


job = JobPosting1('Data Scientist', 'Tech Innovations', 'New York', 120000)

print(job.to_dict())
```

```python
class JobPosting2:
  def __init__(self, title, company, location, salary):
    self.title = title
    self.location = location
    self.company = company
    self.salary = salary

  def compare_salary(self, other_job):
    if self.salary > other_job.salary:
      return f'{self.title} has a higher salary than {other_job.title}'
    else:
        return f'{other_job.title} has a higher salary than {self.title}'
```

```python
job1 = JobPosting2('Data Scientist', 'Tech Innovations', 'New York', 120000)
job2 = JobPosting2('Data Analyst', 'Data Driven Co', 'San Francisco', 100000)

print(job1.compare_salary(job2))
```

```python
# Import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.index
```

```python
df.index.name = 'job_index'
df.head(5)
```

```python
df.index.dtype
```

```python
df_usa = df[df['job_country'] == 'United States']

df_usa.head(5)
```

```python
df_usa.reset_index(inplace=True)
df_usa.head()
```

```python
df_usa.set_index('job_index', inplace=True)

df_usa.head()
```

```python
median_pivot = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending = False)
median_pivot
```

```python
median_pivot.sort_index(inplace=True)
median_pivot
```

```python
df_ger = df[df['job_country'] == 'Germany'].copy()
```

```python
df_ger['job_posted_month'] = df_ger['job_posted_date'].dt.strftime('%B')
df_ger
```

```python
df_ger_pivot = df_ger.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')
```

```python
df_ger_pivot.reset_index(inplace = True)
```

```python
df_ger_pivot['month_no'] = pd.to_datetime(df_ger_pivot['job_posted_month'], format='%B').dt.month
df_ger_pivot.sort_values('month_no', inplace = True)

df_ger_pivot['month_no']
```

```python
df_ger_pivot.set_index('job_posted_month', inplace = True)
```

```python
df_ger_pivot.drop(columns = 'month_no', inplace=True)
```

```python
df_ger_pivot
```

```python
df_ger_pivot.plot(kind = 'line')
```

```python
top_3 = df_ger['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_ger_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in Germany')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show
```
